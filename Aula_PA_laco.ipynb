#Q1
* contagem simples, o usuário digita um número e o programa imprime do zero até o número.

```python
#entrada
cnt = (int(input("Digite o número a ser contado: ")))
for i in range(cnt+1):
  print(i)
```

#Q2
* contagem regressiva, o usuário digita um número e o programa imprime do número até o zero

```python
#Entrada
cnt = (int(input("Digite o número a ser contado: ")))
for i in range(cnt, -1, -1):
  print(i)
```

#Q5
* soma, o usuário digita um número e o programa soma os números do zero até o número

```python
#Entrada
cnt = (int(input("Digite o número a ser contado: ")))
soma = 0
for i in range(cnt+1):
  soma = soma + i
  print(soma)
```

#Q8
* Contar vogais, o usuário digita uma palavra e o programa conta as vogais

```python
#Entrada
texto = input("Digite uma palavra ou frase: ")
vogais = "aeiouAEIOU"
contador = 0
for caractere in texto:
    if caractere in vogais:
        contador += 1

print(f"A palavra/frase '{texto}' contém {contador} vogais.")
```

#Q9
* maior número digitado, o usuário digita quatro números e o programa imprime o maior número digitado

```python
#Entrada
cnt1 = (int(input("Digite o primeiro número: ")))
cnt2 = (int(input("Digite o segundo número: ")))
cnt3 = (int(input("Digite o terceiro número: ")))
cnt4 = (int(input("Digite o quarto número: ")))
maior = cnt1
if cnt2 > maior:
  maior = cnt2
if cnt3 > maior:
  maior = cnt3
if cnt4 > maior:
  maior = cnt4
print(f"O maior número digitado foi {maior}")
```

#Q10
* o usuário digita um número de 1 a 100 e o programa mostra quais são divisiveis por 3 e 5

```python
#Entrada
cnt = (int(input("Digite um número de 1 a 100: ")))
for i in range(1, cnt+1):
  if i % 3 == 0 and i % 5 == 0:
    print (f"{i} é divisível por 3 e 5")
```
